# Crawl web images

```python
# imports
from pathlib                  import Path
from fastai2.basics           import *
from fastai2.vision.all       import *

import sys
sys.path.append(".")
from ai_utilities2 import *
```

### Set root path

```python
root_path = Path("./datasets")
root_path.ls()
```

### Image search parameters

```python
image_dir = root_path/"oak_leaves"
```

```python
labels = ['Northern Red', 'White', 'Chestnut']
search_texts = [f"{l} oak leaves" for l in labels]
```

```python
num_images = 500
engine = "bing"
apikey = None
```

### Run crawler

```python
for i,label in enumerate(labels):
    search_text = search_texts[i]
    image_download(search_text, num_images, label, engine, image_dir, apikey)
```

### Create train/valid/test subfolders?

```python
make_train_valid_flag = False
train, valid = 0.8, 0.2 
test  = 1 - train - valid
```

```python
# split into train/valid/test folders
if make_train_valid_flag:
    path = Path.cwd()/image_dir
    make_train_valid(path, train, valid, test)
```
